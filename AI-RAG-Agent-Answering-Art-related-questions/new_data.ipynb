```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

def image_description(driver, url):
    driver.get(url)
    wait = WebDriverWait(driver, 10)
    element = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "artwork__title--text js-artwork__title--text")))
    title = element.text
    
    element = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/section[2]/div/div/div[1]/div/div/p")))
    
    info = element.text
    description = title + "\n\n" + info
    
    return description
    
def image_urls(driver, num):
    try:
        wait = WebDriverWait(driver, 10)
        link_img_list = []

        while len(link_img_list) < num:
            # Attendre et collecter les liens sur la page actuelle
            elements = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'redundant-link_redundantlink__b5TFR')))
            
            for element in elements:
                if len(link_img_list) >= num:
                    break
                try:
                    link = element.get_attribute("href")
                    if link and link not in link_img_list:
                        link_img_list.append(link)
                except Exception as e:
                    print(f'Erreur lors du traitement d\'un Ã©lÃ©ment : {e}')
                    continue
            
            # VÃ©rifier si le bouton "Next" existe et cliquer pour aller Ã  la page suivante
            try:
                next_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='Next Page']")))
                next_button.click()
                time.sleep(5)  # Pause pour charger la nouvelle page
            except Exception:
                print("Bouton 'Next' introuvable ou derniÃ¨re page atteinte.")
                break
        
        return link_img_list

    except Exception as e:
        print(f'Erreur gÃ©nÃ©rale : {e}')
    finally:
        print('Extraction terminÃ©e.')

# Configuration du navigateur
options = Options()
options.add_argument('--start-maximized')
service = Service('chromedriver.exe')
driver = webdriver.Chrome(service=service, options=options)

# AccÃ©der Ã  la page de recherche du site
driver.get("https://www.metmuseum.org/art/collection/search?q=carthage")
time.sleep(2)    

# RÃ©cupÃ©rer les liens d'images (par exemple, 15 images)
img_links = image_urls(driver, 100)

# Affichage des liens extraits
for i, link in enumerate(img_links, 1):
    print(f"{i}: {link}")

# Fermer le navigateur
driver.quit()
```

```python
import os
import io
import time
import requests
from PIL import Image  
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Configuration du navigateur Selenium
options = Options()
options.add_argument('--start-maximized')
service = Service('chromedriver.exe')
driver = webdriver.Chrome(service=service, options=options)

def create_folders(civilization_name):
    """CrÃ©er les dossiers pour stocker les images et descriptions."""
    base_path = os.path.join(os.getcwd(), civilization_name)
    image_path = os.path.join(base_path, "images")
    description_path = os.path.join(base_path, "descriptions")
    
    os.makedirs(image_path, exist_ok=True)
    os.makedirs(description_path, exist_ok=True)
    
    return image_path, description_path

def get_image_links(driver, search_url, num_images):
    """RÃ©cupÃ¨re les liens des images depuis la page de recherche."""
    driver.get(search_url)
    wait = WebDriverWait(driver, 10)
    image_links = []

    while len(image_links) < num_images:
        try:
            elements = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'redundant-link_redundantlink__b5TFR')))
            for element in elements:
                if len(image_links) >= num_images:
                    break
                link = element.get_attribute("href")
                if link and link not in image_links:
                    image_links.append(link)
        except Exception as e:
            print(f"Erreur lors de l'extraction des liens : {e}")
            break

        # Passage Ã  la page suivante
        try:
            next_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='Next Page']")))
            next_button.click()
            time.sleep(5)  
        except:
            print("Fin de la pagination ou bouton 'Next' introuvable.")
            break

    return image_links

def get_image_description(driver, url):
    """Extrait le titre, la description et l'URL de l'image d'une Å“uvre."""
    try:
        driver.get(url)
        wait = WebDriverWait(driver, 10)

        title = "Titre non trouvÃ©"
        description = "Description non trouvÃ©e"
        img_url = "Image non trouvÃ©e"

        try:
            title_element = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/section[2]/div/div/div[1]/div/h1/span/span")))
            title = title_element.text.strip()
        except:
            pass

        try:
            desc_element = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/section[2]/div/div/div[1]/div/div/p")))
            description = desc_element.text.strip()
        except:
            pass

        try:
            img_element = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/section[2]/div/div/div[2]/div/figure/div[1]/img")))
            img_url = img_element.get_attribute("src")
        except:
            pass

        return {
            "title": title,
            "description": description,
            "image_url": img_url
        }
    except Exception as e:
        print(f"Erreur lors de l'extraction des dÃ©tails : {e}")
        return None

def download_image(download_path, url, file_name):
    """TÃ©lÃ©charge une image et l'enregistre localement."""
    if not url or "Image non trouvÃ©e" in url:
        print(f"âš ï¸ Image introuvable pour {file_name}, tÃ©lÃ©chargement ignorÃ©.")
        return
    
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()
        
        image_file = io.BytesIO(response.content)
        image = Image.open(image_file)
        
        file_path = os.path.join(download_path, file_name)
        image.save(file_path, "JPEG")
        
        print(f"âœ… Image tÃ©lÃ©chargÃ©e : {file_name}")
    except Exception as e:
        print(f"âŒ Ã‰chec du tÃ©lÃ©chargement de {file_name} - Erreur : {e}")

def save_description(description_path, file_name, description_data):
    """Enregistre la description dans un fichier texte."""
    try:
        file_path = os.path.join(description_path, file_name)
        with open(file_path, "w", encoding="utf-8") as f:
            f.write(f"Titre: {description_data['title']}\n\n")
            f.write(f"Description: {description_data['description']}\n")
        
        print(f"ğŸ“„ Description sauvegardÃ©e : {file_name}")
    except Exception as e:
        print(f"Erreur lors de l'enregistrement de la description : {e}")

def main():
    civilization_name = "Carthage"  # Modifier selon la civilisation recherchÃ©e
    search_url = "https://www.metmuseum.org/art/collection/search?q=carthage"
    num_images = 150  # Modifier selon le besoin

    image_path, description_path = create_folders(civilization_name)

    print("ğŸ” Extraction des liens d'images...")
    image_links = get_image_links(driver, search_url, num_images)

    for idx, link in enumerate(image_links):
        print(f"\nğŸ“œ Traitement de l'image {idx+1} sur {len(image_links)}")
        
        data = get_image_description(driver, link)
        if data:
            file_name = f"image_{idx+1}.jpg"
            desc_file_name = f"image_{idx+1}.txt"
            
            download_image(image_path, data["image_url"], file_name)
            save_description(description_path, desc_file_name, data)

    driver.quit()
    print("\nâœ… Processus terminÃ© !")

if __name__ == "__main__":
    main()
```
